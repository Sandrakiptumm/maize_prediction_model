# EXPLORATORY DATA ANALYSIS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("clean_data")
data.head(10)
```

## Univariate data analysis

```python
# Example for Categorical Variables
sns.countplot(x="County", data=data)
plt.xticks(rotation=90) 
plt.show()
```

```python
sns.countplot(x="Classification", data=data)
plt.xticks(rotation=90) 
plt.show()
```

```python
# Calculate the average Wholesale and Retail prices for each Classification
avg_prices = data.groupby("Classification")[["Wholesale", "Retail"]].mean().reset_index()

# Plot average Wholesale prices by Classification
plt.figure(figsize=(10, 6))
sns.barplot(x="Classification", y="Retail", data=avg_prices, color="orange", label="Retail")
sns.barplot(x="Classification", y="Wholesale", data=avg_prices, color="blue", label="Wholesale")


# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add title and labels
plt.title("Average Wholesale and Retail Prices by Classification")
plt.ylabel("Average Price")
plt.xlabel("Classification")

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
data.head(2)
```

```python
sns.histplot(data["Wholesale"], kde=True, color="blue", label="Wholesale")
sns.histplot(data["Retail"], kde=True, color="orange", label="Retail")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Wholesale and Retail Price Distribution")
plt.legend()
plt.show()
```

## Bivariate Data analysis

```python
# Plot average Supply Volume by County
plt.figure(figsize=(10, 6))
sns.barplot(x="County", y="Supply Volume", data=data, estimator='mean', ci=None)
plt.xticks(rotation=90)
plt.title("Average Supply Volume by County")
plt.show()

```

```python
# Plot average Wholesale price by County
plt.figure(figsize=(10, 6))
sns.barplot(x="County", y="Wholesale", data=data, estimator='mean', ci=None)
plt.xticks(rotation=90)
plt.title("Average Wholesale Price by County")
plt.show()

# Plot average Wholesale price over Time
plt.figure(figsize=(10, 6))
sns.lineplot(x="Date", y="Wholesale", data=data, estimator='mean', ci=None)
plt.xticks(rotation=45)
plt.title("Average Wholesale Price over Time")
plt.show()
```

```python
# Plot average Retail price by County
plt.figure(figsize=(10, 6))
sns.barplot(x="County", y="Retail", data=data, estimator='mean', ci=None)
plt.xticks(rotation=90)
plt.title("Average Retail Price by County")
plt.show()

# Plot average Retail price over Time
plt.figure(figsize=(10, 6))
sns.lineplot(x="Date", y="Retail", data=data, estimator='mean', ci=None)
plt.xticks(rotation=45)
plt.title("Average Retail Price over Time")
plt.show()
```

```python
# Plot both Wholesale and Retail prices by County
plt.figure(figsize=(10, 6))
sns.barplot(x="County", y="Wholesale", data=data, estimator='mean', ci=None, color='blue', label="Wholesale")
sns.barplot(x="County", y="Retail", data=data, estimator='mean', ci=None, color='orange', label="Retail")
plt.xticks(rotation=90)
plt.title("Average Wholesale and Retail Prices by County")
plt.legend()
plt.show()

# Plot both Wholesale and Retail prices over Time
plt.figure(figsize=(10, 6))
sns.lineplot(x="Date", y="Wholesale", data=data, estimator='mean', ci=None, label="Wholesale", color='blue')
sns.lineplot(x="Date", y="Retail", data=data, estimator='mean', ci=None, label="Retail", color='orange')
plt.xticks(rotation=45)
plt.title("Average Wholesale and Retail Prices over Time")
plt.legend()
plt.show()
```

## multivariate analysis


```python
numerical_columns = data[["Wholesale", "Retail", "Supply Volume"]]

# Plot the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(numerical_columns.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Wholesale, Retail, and Supply Volume")
plt.show()
```

```python
# Plot pairwise relationships between numerical columns and hue by Classification
sns.pairplot(data, vars=["Wholesale", "Retail", "Supply Volume"], hue="Classification", height=2.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Wholesale", y="Retail", hue="Classification", data=data, palette="Set2")
plt.title("Wholesale vs Retail Prices by Classification")
plt.show()
```

```python
# FacetGrid for Wholesale and Retail prices across different Counties
g = sns.FacetGrid(data, col="County", hue="Classification", height=4, col_wrap=4)
g.map(sns.scatterplot, "Wholesale", "Retail", alpha=0.7)
g.add_legend()
plt.show()
```
