# EXPLORATORY DATA ANALYSIS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("clean_data")
data.head()
```

## Univariate data analysis

```python
# Example for Categorical Variables
sns.countplot(x="County", data=data)
plt.xticks(rotation=90) 
plt.show()
```

```python
sns.histplot(data["Wholesale"], kde=True)
plt.show()
```

```python
sns.histplot(data["Retail"], kde=True)
plt.show()
```

```python
sns.histplot(data["Supply Volume"], kde=True)
plt.show()
```
